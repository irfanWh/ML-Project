```python
# Analyser les colonnes
print("=== ANALYSE DES FEATURES ===\n")

# 1. Valeurs manquantes
print("Valeurs manquantes par colonne:")
print(dataset.isnull().sum())
print("\n")

# 2. Colonnes constantes
print("Colonnes avec une seule valeur unique:")
for col in dataset.columns:
    if dataset[col].nunique() == 1:
        print(f"  - {col}")
print("\n")

# 3. Cardinalité (nombre de valeurs uniques)
print("Cardinalité des colonnes:")
print(dataset.nunique().sort_values())
print("\n")

# 4. Types de données
print("Types de données:")
print(dataset.dtypes)
```

```python
# Colonnes à supprimer
colonnes_a_supprimer = [
    'companyName', 'companyNameMandataire', 'affiliateNumberMandataire',
    'admin_firstName', 'admin_lastName', 'admin_cin', 'admin_email', 'admin_phoneNumber', 'admin_isRL',
    'bank_bankCode', 'bank_adherent_id', 'bank_adherent_numAffilie', 'bank_dateCreation',
    'bank_accountId', 'bank_adherent_raisonSocial', 'bank_accountRIB', 'bank_adherent_adherentMandataire'
]

# Filtrer seulement les colonnes qui existent dans le dataset
colonnes_existantes = [col for col in colonnes_a_supprimer if col in dataset.columns]

# Supprimer les colonnes
dataset_clean = dataset.drop(columns=colonnes_existantes)

print(f"Colonnes supprimées : {colonnes_existantes}")
print(f"\nNombre de colonnes avant : {len(dataset.columns)}")
print(f"Nombre de colonnes après : {len(dataset_clean.columns)}")
print(f"\nColonnes restantes : {list(dataset_clean.columns)}")
```

```python
# Afficher l'état avant nettoyage
print("=== AVANT NETTOYAGE ===")
print(f"Nombre de lignes : {len(dataset_clean)}")
print(f"Nombre de doublons : {dataset_clean.duplicated().sum()}")
print(f"Nombre de lignes avec valeurs nulles : {dataset_clean.isnull().any(axis=1).sum()}")
print("\n")

# 1. Supprimer les doublons
dataset_clean = dataset_clean.drop_duplicates()
print(f"Après suppression des doublons : {len(dataset_clean)} lignes")

# 2. Supprimer les lignes avec valeurs nulles
dataset_clean = dataset_clean.dropna()
print(f"Après suppression des nulles : {len(dataset_clean)} lignes")

# Afficher l'état après nettoyage
print("\n=== APRÈS NETTOYAGE ===")
print(f"Nombre de lignes final : {len(dataset_clean)}")
print(f"Valeurs nulles restantes : {dataset_clean.isnull().sum().sum()}")
print(f"Doublons restants : {dataset_clean.duplicated().sum()}")
```

```python
for i in dataset_clean.columns:
    print(i+':',dataset_clean[i].unique())
```

```python
# Convertir les colonnes de dates en datetime
date_columns = ['dateAdhesion', 'dateAffiliation']

for col in date_columns:
    if col in dataset_clean.columns:
        dataset_clean[col] = pd.to_datetime(dataset_clean[col], errors='coerce')

# Features d'ancienneté (mois depuis aujourd'hui)
if 'dateAdhesion' in dataset_clean.columns:
    dataset_clean['anciennete_adhesion_months'] =(pd.Timestamp.now() - dataset_clean['dateAdhesion']).dt.days / 30.44
    dataset_clean['anciennete_adhesion_months'] = dataset_clean['anciennete_adhesion_months'].round(0)


if 'dateAffiliation' in dataset_clean.columns:
    dataset_clean['anciennete_affiliation_months'] = (pd.Timestamp.now() - dataset_clean['dateAffiliation']).dt.days / 30.44
    dataset_clean['anciennete_affiliation_months'] = dataset_clean['anciennete_affiliation_months'].round(0)

# Différence entre les deux dates
if 'dateAdhesion' in dataset_clean.columns and 'dateAffiliation' in dataset_clean.columns:
    dataset_clean['months_adhesion_affiliation'] = (dataset_clean['dateAdhesion'] - dataset_clean['dateAffiliation']).dt.days / 30.44
    dataset_clean['months_adhesion_affiliation'] = dataset_clean['months_adhesion_affiliation'].round(0)

# Supprimer les colonnes de dates originales
dataset_clean = dataset_clean.drop(columns=[col for col in date_columns if col in dataset_clean.columns])

# Afficher les nouvelles features extraites
print("Features extraites des dates:")
print([col for col in dataset_clean.columns 
       if any(keyword in col for keyword in ['adhesion', 'affiliation'])])
```

```python
dataset_clean['anciennete_adhesion_months'] = dataset_clean['anciennete_adhesion_months'].astype(int)
dataset_clean['anciennete_affiliation_months'] = dataset_clean['anciennete_affiliation_months'].astype(int)
dataset_clean['months_adhesion_affiliation'] = dataset_clean['months_adhesion_affiliation'].astype(int)
```

## Étape 1: Encoder les variables catégorielles

```python
# Identifier les colonnes catégorielles (type object)
categorical_cols = dataset_clean.select_dtypes(include=['object']).columns.tolist()
print("Colonnes catégorielles à encoder:")
for col in categorical_cols:
    print(f"  - {col}: {dataset_clean[col].nunique()} valeurs uniques")
print(f"\nTotal: {len(categorical_cols)} colonnes")
```

```python
from sklearn.preprocessing import LabelEncoder

# Créer une copie du dataset
dataset_encoded = dataset_clean.copy()

# Identifier les colonnes catégorielles (type object)
categorical_cols = dataset_encoded.select_dtypes(include=['object']).columns.tolist()

print("=== ENCODAGE DES VARIABLES CATÉGORIELLES ===\n")
print(f"Colonnes à encoder : {categorical_cols}\n")

# Encoder chaque colonne catégorielle
label_encoders = {}
for col in categorical_cols:
    if col == 'affiliateNumber':
        continue
    le = LabelEncoder()
    dataset_encoded[col] = le.fit_transform(dataset_encoded[col].astype(str))
    label_encoders[col] = le
    print(f"✓ {col}: {len(le.classes_)} classes → encodé en [0-{len(le.classes_)-1}]")

print(f"\n✓ Encodage terminé pour {len(categorical_cols)} colonnes")

# Vérifier le résultat
print("\n=== VÉRIFICATION ===")
print(f"Colonnes de type 'object' restantes : {len(dataset_encoded.select_dtypes(include=['object']).columns)}")
print(f"\nTypes de données après encodage :")
print(dataset_encoded.dtypes.value_counts())

# Afficher un aperçu
print("\nAperçu des données encodées :")
dataset_encoded.head()
```

```python
# Supprimer les doublons en gardant la ligne avec la plus grande ancienneté
print("=== SUPPRESSION DES DOUBLONS PAR affiliateNumber ===\n")
print(f"Nombre de lignes avant : {len(dataset_encoded)}")

# Trier par affiliateNumber et anciennete_adhesion_months (décroissant)
dataset_encoded = dataset_encoded.sort_values('anciennete_adhesion_months', ascending=False)

# Garder la première ligne de chaque affiliateNumber (= celle avec la plus grande ancienneté)
dataset_encoded = dataset_encoded.drop_duplicates(subset=['affiliateNumber'], keep='first')

print(f"✓ Nombre de lignes après : {len(dataset_encoded)}")
print(f"✓ Gardé : la ligne avec la plus grande ancienneté pour chaque adhérent")
```

```python
import pickle

# 1. Sauvegarder le dataset encodé
dataset_encoded.to_csv('dataset_encoded_adherents.csv', index=False)

# 2. Sauvegarder les label encoders (pour pouvoir décoder)
with open('label_encoders_adherents.pkl', 'wb') as f:
    pickle.dump(label_encoders, f)

# 3. Sauvegarder aussi les mappings en CSV lisible
mappings_df = pd.DataFrame()
for col in categorical_cols:
    mapping = pd.DataFrame({
        'column': col,
        'code': range(len(label_encoders[col].classes_)),
        'original_value': label_encoders[col].classes_
    })
    mappings_df = pd.concat([mappings_df, mapping], ignore_index=True)

mappings_df.to_csv('encodage_mappings_adherents.csv', index=False)

print("✓ Fichiers sauvegardés :")
print("  - dataset_encoded_adherents.csv")
print("  - label_encoders_adherents.pkl")
print("  - encodage_mappings_adherents.csv")
```
